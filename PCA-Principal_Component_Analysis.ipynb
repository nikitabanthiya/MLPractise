### Dimensionality Reduction


#### PCA is a dimensionality reduction technique; 
#### it lets you distill multi-dimensional data down to fewer dimensions, 
#### selecting new dimensions that preserve variance in the data as best it can.


```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import pylab as pl
from itertools import cycle
```

```python
iris = load_iris()

numSamples, numFeatures = iris.data.shape
print(numSamples)
print(numFeatures)
print(list(iris.target_names))
```

```python
X = iris.data
pca = PCA(n_components=2, whiten=True).fit(X)
X_pca = pca.transform(X)
```

```python
print(pca.components_)
```

```python
print(pca.explained_variance_ratio_)
print(sum(pca.explained_variance_ratio_))
```

```python
%matplotlib inline
from pylab import *

colors = cycle('rgb')
target_ids = range(len(iris.target_names))
pl.figure()
for i, c, label in zip(target_ids, colors, iris.target_names):
    pl.scatter(X_pca[iris.target == i, 0], X_pca[iris.target == i, 1],
        c=c, label=label)
pl.legend()
pl.show()
```

```python
pca_1 = PCA(n_components=1, whiten=True).fit(X)
X_pca_1 = pca_1.transform(X)
```

```python
print(pca_1.components_)
```
